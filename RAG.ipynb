# Storing

```python
import chromadb
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core import StorageContext
```

```python
db = chromadb.PersistentClient(path="./chroma_db")
```

```python

chroma_collection = db.get_or_create_collection("quickstart")
```

```python
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
```

```python
index = VectorStoreIndex.from_documents(
    data , storage_context=storage_context
)
```

```python
index
```

# Querying

```python
text = """Generate questions for the topic Pompe disease using Bloom's Taxonomy. For each level, create the following number of questions:

Remember: [0] questions that focus on recalling facts, terms, or basic concepts.
Understand: [0] questions that assess comprehension and interpretation of the topic.
Apply: [0] questions that require applying knowledge in practical or real-world scenarios.
Analyze: [0] questions to break down information, examine relationships, or draw comparisons.
Evaluate: [0] questions that ask learners to form judgments, justify opinions, or evaluate evidence.
Create: [0] questions that encourage producing new ideas, designs, or original work based on the topic.
Each question should align with the cognitive process relevant to its Bloom's Taxonomy level."""
```

# Result

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query(text)
```
